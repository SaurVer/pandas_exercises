### Step 4. Delete the Unnamed 0 column

### Step 5. Plot the total_bill column histogram

```python
sns.distplot(tips.total_bill)
plt.title('total bill frequency')
plt.xlabel('total_bill')
plt.ylabel('frequency')
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python

ax=sns.jointplot(x=tips.total_bill, y= tips.tip)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a FacetGrid with one joint plot per day
g = sns.FacetGrid(tips, col='day', col_wrap=2)
g.map(sns.scatterplot, 'total_bill', 'tip')

# Set labels and title for each subplot
for ax in g.axes.flat:
    ax.set_xlabel('Total Bill')
    ax.set_ylabel('Tip')
    # ax.set_title(ax.get_title().split('=')[-1])  # Extract day from subplot title

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
