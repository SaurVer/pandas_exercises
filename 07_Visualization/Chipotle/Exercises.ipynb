### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo= pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv')
chipo
```

### Step 4. See the first 10 entries

```python
chipo.head(10)
```

### Step 5. Create a histogram of the top 5 items bought

```python
top_items=chipo.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(5).to_dict()
```

```python
fig= plt.figure(figsize=(10,5))
plt.bar(top_items.keys(), top_items.values(),color= 'cyan')
plt.title('Top items sold')
plt.xlabel('items')
plt.ylabel('number of items sold')

plt.xticks(rotation= 25)
plt.tight_layout()
plt.legend()
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
items_order_for_price=chipo.groupby('item_price')['quantity'].sum().sort_values(ascending=False).to_dict()
items_order_for_price
```

```python
plt.figure(figsize=(25,5))
plt.scatter(x=items_order_for_price.keys(), y= items_order_for_price.values(), color= 'y')
plt.xticks(rotation=90)
plt.title(' items sold')
plt.xlabel('price point')
plt.ylabel('number of items sold')
```
