### Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.

```python
import pandas as pd
from datetime import datetime

# Read the dataset from the URL
data = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data', sep="\s+")

# Create a function to convert 'Yr', 'Mo', and 'Dy' to datetime objects
def combine_date(row):
    return datetime(int(row['Yr']), int(row['Mo']), int(row['Dy']))

# Apply the function to each row to create the 'Date' column
data['Date'] = data.apply(combine_date, axis=1)

# Display the DataFrame with the new 'Date' column
print(data)
```
